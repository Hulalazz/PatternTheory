# 1. Objective
This project is aimed at using Gaussian models to simulate sounds generated from real instruments. 

In a synthesizer, we cannot merely play recorded sound as there are many factors involved during spontaneous playing. 
Take for example the piano. 
The sound is different depending on how quickly and deep the key is pressed and when it is released. 
To record all possible combinations of sound costs too much memory. Therefore modern electronic pianos generate the notes on the fly.

Different instruments have different sounds, so each of these instruments have a distinct template. The template consist of information such as harmonics height and width, ADSR  envelope, inharmonicity coefficient. Additionally, there are also additional variables that affect the eventual sound: duration, time points ADSR, etc. 

### 1.1 ADSR model
Most physical instruments conform to the ADSR model: https://en.wikipedia.org/wiki/Synthesizer#Attack_Decay_Sustain_Release_.28ADSR.29_envelope
> The contour of an ADSR envelope is specified using four parameters:
Attack time is the time taken for initial run-up of level from nil to peak, beginning when the key is first pressed.
Decay time is the time taken for the subsequent run down from the attack level to the designated sustain level.
Sustain level is the level during the main sequence of the sound's duration, until the key is released.
Release time is the time taken for the level to decay from the sustain level to zero after the key is released.

## Calculating time envelope

Directly below the frequency spectrum, we then see the signal $s(t)$ in the time domain (in blue) and our generated envelope (in green) that links the peaks of the reference together. 


The signal is made up of values, representing displacement along the y-axis, sampled at the sampling rate (44100 Hz). 
In our project, we standardise the duration of the note to be 1 second.
The code is provided below, with the input being a list of the audio file displacement values in each instance of time (44100 Hz). 

In other words, the frequency (pitch) of the note can be quantified via the signal $s(t)$ of the wave and remains constant, while the amplitudes of $s(t)$ represent the loudness of the note at various points in time.
Following is a piece of code which approximates the envelope of a given signal. 
This code will also be used for all subsequent occasions when a window or a graph is to be generated.



```python
def calculate_time_profile(audio, zero_pad=False):
    N = len(audio)
    num_blocks = 100
    # number of blocks is arbitrary chosen. Ideally, the ffreq should be larger than the number of blocks.
    block_size = int(N / num_blocks)

    peaks = []
    for block in range(num_blocks):  # measuring the peaks
        maxi = max(audio[block * block_size:(block + 1) * block_size])
        mini = min(audio[block * block_size:(block + 1) * block_size])
        peaks.append(maxi / 2 - mini / 2)

    window = [0] * N
    for x in range(num_blocks - 1):  # connecting the peaks together
        for y in range(block_size):
            window[x * block_size + y + int(block_size / 2)] = peaks[x] + (peaks[x + 1] - peaks[x]) / block_size * y
    for y in range(int(block_size / 2 + 1)):
        window[N - y - 1] = window[N - block_size / 2 - 2]
        window[y] = window[block_size / 2]
    # to avoid dividing by zero in normalise_throughout
    if zero_pad:
        window = 0.99 * np.array(window) + [0.01] * N
    return window
```

```python
time_envelope = hp.calculate_time_envelope(ref)

plt.figure(figsize=(16,4))
plt.plot(ref)
plt.plot(time_envelope)
plt.show()
```

## Introducing random phases to a frequency spectrum

We then added (uniform) random phase to each frequency, producing the waveform below, which we will call "sample". 
In other words, we took the waveform in the frequency domain above and gave them random phase (starting at random times) in the time domain.
The code used is displayed below the graph and the random phase is introduced in the complete_magnitude function.

```python
plt.figure(figsize=(16,4))
plt.plot(signal_fft.real[:2048])
plt.plot(signal_fft.imag[:2048])
plt.plot(np.absolute(signal_fft)[:2048])
plt.show()
```

```python
def complete_magnitude(magnitude):
    # the input is the absolute value, excludes zero at 0 - total of N/2 - 1
    phase = [2 * np.pi * np.random.random() for angle in range(len(magnitude))]
    reals = np.multiply(magnitude, np.sin(np.array(phase)))
    imags = np.multiply(magnitude, np.cos(np.array(phase)))

    fcef = complete_realimags(reals, imags)
    return fcef


def complete_realimags(reals, imags):
    # the input excludes zero at 0 - total of N/2 - 1 entries
    reals = np.concatenate([reals, reals[::-1]])
    reals = np.insert(reals, int(len(reals) / 2), 0)
    reals = np.insert(reals, 0, 0)

    imags = np.concatenate([imags, -1 * np.array(imags[::-1])])
    imags = np.insert(imags, int(len(imags) / 2), 0)
    imags = np.insert(imags, 0, 0)

    fcef = np.vectorize(complex)(reals, imags)
    return fcef


def ifft(fcef):
    signal = np.fft.ifft(fcef)  # different from the library implmenetation as it discard imaginary part
    signal_real = [entry.real for entry in signal]
    return signal_real
```

### Envelop fitting

For the sound to sound like an instrument, for example the piano, it has to have a high amplitude at the start (Attack/Decay) and a gradually decreasing amplitude (Sustain/Release). 
To achieve the realistic effect with our generated waveform, we first obtained a "even" profile by using the function normalise\_through, which divides the sample audio by our calculated time profile (of itself). 
A side note is also that, to avoid dividing by zero, we had added a little padding to the time profile. The code and waveform with even amplitude after this process is shown below.

```python
def normalise_through(audio):
    sample_time_profile = calculate_time_profile(audio, True)
    audio = np.divide(audio, sample_time_profile)
    return audio
```

```python
sample_normalised = hp.normalise_through(sample)
plt.figure(figsize=(16,4))
plt.plot(sample_normalised)
plt.show()
```

```python
sd.play(sample_normalised, 44100)  # play sound
```

We tested this process out with 8 different sounds. We realised that this process affected several sounds more significantly. 
Those sounds had only 1 or 2 harmonics, and are very uneven in the time domain. In other words, constructive and destructive interference affected them more significantly.  \\

We then multiply the result with the time profile of the reference (the one in green):  

```python
def envelope_fitting(audio, ref_time_envelope):
    audio = normalise_through(audio)
    audio = np.multiply(audio, ref_time_envelope)
    return audio
```

```python
sample_normalised_fitted = hp.envelope_fitting(sample_normalised, time_envelope)
plt.figure(figsize=(16,4))
plt.plot(sample_normalised_fitted)
plt.show()
```

```python
sd.play(sample_normalised_fitted, 44100)  # play sound
```

# 5. Sampling in time using the Gaussian Model

> (Taken from page 76 of Pattern Theory) Let us take a discrete sample of the signal $s$ and, for simplicity, lets assume that $s$ wraps around at some large integer $N$ (i.e. $s_{N+k} = s_k$) and that period $p$ is an integer dividing $N$. Let $q = N/p$, the number of cycles present in the whole sample. Well now analyse the simplest possible Gaussian model for $s$ that gives samples that are periodic plus some small residual noise. Its probability density is 
$$p_{a,b}(s) = \frac{1}{Z} e^{-a \sum_{k=0}^{N-1} {(s(k) - s(k+p))2/2} - b \sum_{k=0}^{N-1} {s(k)^2/2}} = \frac{1}{Z} e^{- \vec{s}^{\,t} Q \vec{s}/2}$$
where parameters $a, b > 0$, so that $Q_{i,i} = b + 2a$ and $Q_{i,i \pm p} = - a$ for $0 \leq i \leq N - 1$ and otherwise 0. Note that $Q$ is the inverse covariance matrix.

In the middle section of the equation, the term with coefficient $a$ is to enforce periodicity, and the term with coefficient $b$ is to enforce an upper limit on the magnitude of the numbers, for a realistic effect.

This is an off-the-shelf method to sample a multivariate normal distribution, from a covariance matrix $C = Q^{-1}$:

```python
# np.random.multivariate_normal(mean, cov[, size])
```

This is the overview of the method of this chapter to imitate the recorded piano note "reference".


![title](images/multivar_gaussian_time_sampling_efficient.png)

As the inverse covariance (as well as its covariance) matrix is circulant, the computation time required to sample in the multivariate Gaussian can be reduced (https://math.stackexchange.com/questions/2234019/fastest-way-of-sampling-multivariable-guassian-with-covariance-matrix-that-is-ci) from $O(N^3)$ to $O(N^2)$. This means that the iterations needed is a square function of size, instead of a cubic one.

After sampling the multivariate Gaussian distribution, we will do post processing on the signal to produce our imitation of the signal. Post processing includes eliminating higher frequencies and to fit the signal into the time envelope of the reference.

## Different values of a and b
For illustration, following are samples of the multivariate Gaussian distribution. As can be observed below, setting $a$ to a non-zero number introduces periodicity within the frequency spectrum. Setting $b$ to a smaller number also introduces less extreme magnitudes.

```python
import samp_inverse_circulant as ic
from scipy.linalg import circulant

def multi_var_gaussian_sampling(N, p, a, b):
    icovx = [0] * N
    icovx[0] = b + 2 * a
    icovx[p] = -a
    icovx[-p] = -a
        
    covx = ic.inv_circulant(icovx)
    sample = np.random.multivariate_normal([0]*N, circulant(covx))
    
    sample_fft = np.absolute(np.fft.fft(sample))

    plt.figure(figsize=(16,4))
    plt.plot(sample)
    plt.show()
    
    plt.figure(figsize=(16,4))
    plt.plot(sample_fft)
    plt.show()
```

## Inverse of a circulant matrix
A circulant matrix is a square matrix where each row vector is rotated one element to the right relative to the preceding row vector. First the covariance matrix $C$ is calculated from the inverse covariance matrix $Q$. 
As the both matrices are circulant, it can be represented by its first row (or column, since we are concerned with symmetric matrices) $\vec{c}$ and $\vec{q}$ respectively. 
The formula for inverting the covariance matrix using $\vec{c}$ and $\vec{q}$ is given below, as well as the corresponding code.

$\mathbb{F}(\vec{x})$ and $\mathbb{F}^{-1}(\vec{x})$ are discrete Fourier transforms and inverse discrete Fourier transforms of an arbitrary complex vector $\vec{x}$ respectively.

https://math.stackexchange.com/questions/886241/inverse-of-a-circulant-matrix

https://stats.stackexchange.com/questions/10795/how-to-interpret-an-inverse-covariance-or-precision-matrix/52886

$$\vec{c} = \mathbb{F}^{-1} \left(\frac{1}{[\mathbb{F}(\vec{q})]_1}, \frac{1}{[\mathbb{F}(\vec{q})]_2}, ... , \frac{1}{[\mathbb{F}(\vec{q})]_N} \right)$$

```python
N = 2000
a = 1
b = 0.01
p = 40

icovx = [0] * N
icovx[0] = b + 2 * a
icovx[p] = -a
icovx[-p] = -a

def inv_circulant(col):
    trans = np.fft.fft(col)
    trans_rcp = 1. / np.array(trans)
    inv = np.fft.ifft(trans_rcp)
    return inv.real

covx = ic.inv_circulant(icovx)

plt.figure(figsize=(16,4))
plt.plot(icovx)
plt.plot(covx,lw=0.5)
plt.show()
```

## Optimised Cholesky factorisation for circulant matrices

Then the Cholesky decomposition is carried out on the covariance matrix to produce a lower triangular matrix.

$$C = L L^{T}$$

With the lower triangular matrix, the multivariate Gaussian can be sampled at $O(N^2)$ time. 

 A Toeplitz matrix or diagonal-constant matrix is a matrix in which each descending diagonal from left to right is constant. Covariance matrices are Toeplitz matrices, although Toeplitz matrices are more general because they do not require the matrix to be symmetric, or circulant.  Below, we show the optimised code to factorise Toeplitz matrices into Cholesky factors. 
 It is translated http://people.sc.fsu.edu/~jburkardt/cpp_src/toeplitz_cholesky/toeplitz_cholesky.html from C++. This algorithm requires the main diagonal to be one, which can be achieved by scaling the entire matrix down or up.

```python
def toeplitz_cholesky_lower(n, col):
    arr = [[0] * n for _ in range(n)]
    g = [[0] * n for _ in range(2)]

    g[0][:] = col  # G[0][0:N] contains the first row.
    g[1][:] = col  # G[1][1:N] contains the first column.

    if g[0][0] != 1:
        print("error predicted, diagonal needs to be 1")

    for entry in range(n):
        arr[entry][0] = g[0][entry]
    g[0][1:] = g[0][:-1]
    g[0][0] = 0.0

    for i in range(1,n):
        rho = - g[1][i]/g[0][i]
        div = np.sqrt((1.0 - rho)*(1.0 + rho))
        for j in range(i,n):
            g1j = g[0][j]
            g2j = g[1][j]
            g[0][j] = (g1j + rho*g2j) / div
            g[1][j] = (rho*g1j + g2j) / div
        for j in range(i,n):
            arr[j][i] = g[0][j]
        for j in range(n-1, i, -1):
            g[0][j] = g[0][j-1]
        g[0][i] = 0.0
    return arr
```

```python
L = toeplitz_cholesky_lower(N, np.array(covx)/covx[0])

# import numpy as np
# import matplotlib.pyplot as plt
import matplotlib.cm as cm

# x,y,temp = np.loadtxt('data.txt').T #Transposed for easier unpacking
# nrows, ncols = 100, 100
# grid = temp.reshape((nrows, ncols))
plt.figure(figsize=(14,14))
plt.imshow(L, cmap=cm.Greys)
plt.colorbar(shrink=0.8)
plt.show()
```

We use the lower factor of the Cholesky decomposition $L$ to sample the multivariate Gaussian distribution. It is done by multiplying $L$ with a standard normal vector (with independent elements). 

```python
sample = np.dot(L, np.random.randn(N))

plt.figure(figsize=(16,4))
plt.plot(sample)
plt.show()

plt.figure(figsize=(16,4))
plt.plot(np.absolute(np.fft.fft(sample)))
plt.show()
```

By multiplying the above Cholesky matrix with the vector containing an independent set of standard normal variables, larger multivariate Gaussian distributions can be sampled faster. 
The resultant waveforms in time and frequency domain using this method can be seen above.
Matrix of size 6000 takes 10 seconds to sample, compared to 1 minute from before.

## Suppression of higher frequencies
In addition, FFT-based filters are used to suppress the higher frequencies. The filter can be applied in the frequency domain by multiplying the spectrum with a maunual calibration curve. The same filter can also be applied in the time-domain using the convolve function, and this is preferable to the former method because the suppression of higher frequencies can then be done simultaneously with the note production. 

```python
def convolve(signal, N):
    modulation = [(x / 10.) * np.exp(-x / 200.) for x in range(1, int(N/2))]
    filter_ = complete_realimags(modulation, [0] * int(N/2 - 1))
    filter_x = np.fft.ifft(filter_)
    signal = np.real(np.fft.ifft(np.fft.fft(signal) * np.fft.fft(filter_x)))
    #signal = np.convolve(filter_x, signal, mode="same")
    return signal
```

# 6. Sampling in the frequency domain
> (From page 77 of Pattern Theory)
Then the expected power at frequency $l$ is the mean of $\left|\hat{s}(l)\right|^2$, which works out to be $$E\left(\left|\hat{s}(l)\right|^2\right) = \frac{1}{b + 4a \, {\sin}^2({\pi pl / N})}$$

## Frequency Modulation
Frequency modulation can be done with a arbitrary defined curve:

```python
def frequency_modulation(spectrum):
    # modulation = [np.arctan(x/100.) * np.exp(-x/200.) for x in range(1,n)]
    modulation = freq_profile
    spectrum = np.multiply(np.abs(spectrum), modulation)
    return spectrum

freq_profile = [(x / 10.) * np.exp(-x / 400.) for x in range(1, int(N/2))]
spectrum = frequency_modulation(spectrum)
    
plt.figure(figsize=(16,4))
plt.plot(spectrum)
plt.show()
```

### Using the spectrum of the reference

Alternatively, frequency modulation can be done with audio reference audio. The magnitude of the frequency is measured for each harmonic.

```python
def calc_freq_profile(spectrum_ref, ffreq, ihmrc=0):
    # requires knowledge of the fundamental frequency, rather specific to middle C
    # inharmonicity is ignored
    peaks = [0]
    for x in range(1, int(N/(ffreq*2) - 1)):
        peaks.append(max(spectrum_ref[int(x*ffreq)-100:int(x*ffreq)+100]))

    freq_profile = [0] * int(N/2)
    for x in range(int(N / (ffreq * 2) - 2)):
        for y in range(int(ffreq) + 1):
            freq_profile[int(x * ffreq) + y] = peaks[x] + (y * (peaks[x + 1] - peaks[x]) / ffreq)
    return freq_profile[:int(N/2)-1]

freq_profile = calc_freq_profile(spectrum, ffreq)

plt.figure(figsize=(16,4))
plt.plot(spectrum)
plt.plot(freq_profile)
plt.show()
```
