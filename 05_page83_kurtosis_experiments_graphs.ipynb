```python
from scipy.stats import kurtosis
import numpy as np
import matplotlib.pyplot as plt
```

## The formula we are trying to show here is:
$X \sim Po(2^m)$

$Y_i \sim Po(1)$

$X = Y_1 + Y_2 + Y_3 ... + Y_{2^m}$


$(kurtosis(Y) - 3) = 1/\lambda = 1$

$(kurtosis(X) - 3) = 1/2^m$

```python
kurtosis_array_X = []
kurtosis_for_size_n = []

upper_range = 10
number_for_experiments = 100  # i.e. number of dots

# for _ in range(number_for_experiments):
#     kurtosis_array_X = []
    
#     for m in range(10):
#         X = np.random.binomial(2**m, 0.4, 100)
#         kurtosis_array_X.append(kurtosis(X))
        
#     plt.scatter(np.arange(10) + 0.1*np.random.randn(10),
#                 kurtosis_array_X, 
#                 s = [0.1]*10)
    
for m in range(10):
    
    kurtosis_array_X = []  # need to reset every time
    
    for _ in range(number_for_experiments):
        
        X = np.random.binomial(2**m, 0.5, 10000)
#         X = np.random.poisson(2**m, 10000)  # create an array of Y
        # 100 samples of 10000 of samples of X
        kurtosis_array_X.append(kurtosis(X))
        
    plt.scatter(m + 0.1*np.random.randn(number_for_experiments), 
                kurtosis_array_X, 
                s = [0.1]*number_for_experiments) # s is the dot size, not important
    kurtosis_for_size_n.append(np.mean(kurtosis_array_X, dtype=np.float32))
    
ax = plt.gca() 
ax.set_xlim([-1, upper_range])
plt.xlabel("log2(n)")
plt.ylabel("excess kurtosis of X")
plt.plot(kurtosis_for_size_n)
plt.show()

# for possion
print(np.log2(-np.array(kurtosis_for_size_n)))
plt.plot(np.log2(-np.array(kurtosis_for_size_n)))
plt.show()
# graph is y=-x    
```
