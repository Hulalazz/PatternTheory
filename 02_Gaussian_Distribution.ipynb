### Normal distribution with different mean

```python
mean = 1
samples = np.random.randn(50) + mean
```

```python
plt.figure(figsize=(16,3))
plt.scatter(samples,[0]*len(samples))
plt.xlim([-5,5])
plt.title("sampling normal distribution with different mean")
plt.show()
```

TODO: Insert formula here

```python
plt.figure(figsize=(16,3))
plt.plot(np.arange(-5,5,0.01),[1/np.sqrt(2) * np.exp(-((x-mean)**2)) for x in np.arange(-5,5,0.01)])
# plot histogram here
plt.xlim([-5,5])
plt.title("sampling standard normal distribution ")
plt.show()
```

### Normal distribution with different mean and variance

```python
mean = 1
std_dev = 2
samples = std_dev*np.random.randn(50) + mean
print samples
```

```python
plt.figure(figsize=(16,3))
plt.scatter(samples,[0]*len(samples))
plt.xlim([-5,5])
plt.title("sampling normal distribution with different mean and variance")
plt.show()
```

TODO: Insert formula here

```python
plt.figure(figsize=(16,3))
plt.plot(np.arange(-5,5,0.01),[1/np.sqrt(2*std_dev) * np.exp(-((x-mean)/std_dev)**2) for x in np.arange(-5,5,0.01)])
# plot histogram here
plt.xlim([-5,5])
plt.title("pdf of normal distribution with different mean and variance")
plt.show()
```
